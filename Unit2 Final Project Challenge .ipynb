```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("SeaTemp_SeaLevel_anomaly.csv")
```

```python
year = df["Year"]
temp = df["Sea temperature anomaly (C)"]
mean = df["Change in mean sea level (mm)"]
```

```python
plt.xlabel("Year")
plt.plot(year,temp, c = "red")
plt.ylabel("Sea temperature anomaly (C)")
plt.legend()
plt.twinx()
plt.plot(year,mean, c = "blue")
plt.ylabel("Change in mean sea level (mm)")
plt.title("Sea Temp and Sea Level Anomaly")
plt.legend(loc = 4)
plt.show()
plt.savefig("Sea Temp and Sea Level Anomaly Graph",dpi = 1000)
```
